```python
#for setting limits to axis
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 7, 18]

plt.plot(x, y, marker='o', linestyle='-')

plt.xlim(2, 4)  # Display data points only between x = 2 and x = 4
plt.ylim(8, 16)  # Display data points only between y = 8 and y = 16

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customizing Axis Limits in Matplotlib')

plt.show()
plt.savefig("graph.png")
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y1 = [10, 15, 13, 7, 18]
y2 = [8, 12, 10, 5, 16]

plt.plot(x, y1, linestyle='-', color='blue', label='Line 1')
plt.plot(x, y2, linestyle='-', color='green', label='Line 2')

plt.fill_between(x, y1, y2)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Fill Between Lines in Matplotlib')

plt.legend()
plt.show()

```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 7, 18]

plt.plot(x, y, marker='o', linestyle='-')

# Customize x-axis tick positions and labels
x_ticks = [1, 2, 3, 4, 5]
x_labels = ['A', 'B', 'C', 'D', 'E']
plt.xticks(x_ticks, x_labels)

# Customize y-axis tick positions and labels
y_ticks = [5, 10, 15, 20]
y_labels = ['Low', 'Average', 'High', 'Very High']
plt.yticks(y_ticks, y_labels)

plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Customizing Tick Positions and Labels in Matplotlib')

plt.show()
```
